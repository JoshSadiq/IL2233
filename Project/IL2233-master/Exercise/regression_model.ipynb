```python
# Import
from sklearn.datasets import fetch_california_housing

# Data management
data = fetch_california_housing()
x,y = data.data, data.target

# Regression problem: Output is continuous value
# -> Linear activation at output layer
# -> Prediction is never perfect for regression, use loss metric (MSE, MAE, RMSE) to determine how far from truth

# Traditional design
# Note 8 neurons input layer (8 traits in california housing dataset)
# Layers have pyramid structure on neurons - Go from high to single output neuron
import torch.nn as nn

seq_model = nn.Sequential(
            nn.Linear(8, 24),
            nn.ReLU(),
            nn.Linear(24, 12),
            nn.ReLU(),
            nn.Linear(12, 6),
            nn.ReLU(),
            nn.Linear(6, 1)
)

# Loss function and optimizer
# Use MSE and Adam (common nn) optimizer
import torch.optim as optim
loss_fn = nn.MSELoss()
optimizer = optim.Adam(seq_model.parameters(), lr = 1e-3)

# Splitting data into training and test sets
import copy
import numpy as np
import torch
import tqdm
from sklearn.model_selection import train_test_split

# train-test split of the dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.7, shuffle=True)
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32).reshape(-1, 1)

# training parameters
n_epochs = 100   # number of epochs to run
batch_size = 10  # size of each batch
batch_start = torch.arange(0, len(X_train), batch_size)

# Hold the best model
best_mse = np.inf   # init to infinity
best_weights = None
history = []

# training loop
for epoch in range(n_epochs):
    seq_model.train()

    with tqdm.tqdm(batch_start, unit="batch", mininterval=0, disable=True) as bar:
        bar.set_description(f"Epoch {epoch}")
        for start in bar:
            # take a batch
            X_batch = X_train[start:start+batch_size]
            y_batch = y_train[start:start+batch_size]

            # forward pass
            y_pred = seq_model(X_batch)
            loss = loss_fn(y_pred, y_batch)

            # backward pass
            optimizer.zero_grad()
            loss.backward()

            # update weights
            optimizer.step()

            # print progress
            bar.set_postfix(mse=float(loss))
            
    # evaluate accuracy at end of each epoch
    seq_model.eval()
    y_pred = seq_model(X_test)
    mse = loss_fn(y_pred, y_test)
    mse = float(mse)
    history.append(mse)
    if mse < best_mse:
        best_mse = mse
        best_weights = copy.deepcopy(seq_model.state_dict())
 
# restore model and return best accuracy
seq_model.load_state_dict(best_weights)
```

```python
import matplotlib.pyplot as plt

# Print history of MSE over epochs
print("MSE: %.2f" % best_mse)
print("RMSE: %.2f" % np.sqrt(best_mse))
plt.plot(history)
plt.show()
```
